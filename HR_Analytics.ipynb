```python
#This code snippet is creating and displaying a heatmap to visualize the correlation between numerical features in the dataset
fig, ax= plt.subplots(figsize=(20,20))
data_corr= data.drop(['Attrition','Department','BusinessTravel','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime'], axis=1)
sns.heatmap(data_corr.corr(),annot=True, linewidths=0.5, fmt='0.1f', ax=ax)
```

```python
#distribution of overtime for employees
sns.countplot(x='OverTime', data=data,palette=['red', 'green'])
fig= plt.gcf()
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.title('Over Time')
plt.show()
```

```python
# distribution of education levels across the company
sns.countplot(data=data, x="Education",palette=['red', 'green','blue','orange','yellow'])
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.title('Education Level')
plt.show()
```

```python
#a bar plot that shows the relationship between 'Attrition' and 'Education
def plot_distribution_by_category(data, col1, col2):
  """
  Plots the distribution of a categorical column (col1) based on
  another categorical column (col2) using seaborn's countplot.

  Args:
    data: The pandas DataFrame containing the data.
    col1: The name of the column to be plotted on the x-axis.
    col2: The name of the column to be used for hue (category separation).
  """
  plt.figure(figsize=(10, 6))  # Adjust figure size as needed
  sns.countplot(data=data, x=col1, hue=col2, palette='viridis')
  plt.title(f'Distribution of {col1} by {col2}')
  plt.xlabel(col1)
  plt.ylabel('Count')
  plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels if needed
  plt.show()

# Now you can call the function:
plot_distribution_by_category(data, col1='Attrition', col2='Education')
```

```python
# Distibution of Attrition
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar Chart
sns.countplot(data, x='Attrition', palette='coolwarm', ax=axes[0])
for bars in axes[0].containers:
    axes[0].bar_label(bars, fmt='%d', label_type='edge', padding=3)
axes[0].set_title('Attrition Count', fontsize=12)
axes[0].set_xlabel('Attrition', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)

# Pie Chart
attrition_counts = data['Attrition'].value_counts()
axes[1].pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("coolwarm", 2))
axes[1].set_title('Attrition Distribution')
axes[1].axis('equal')
plt.tight_layout()
plt.show()
```

```python

# Calculate quartiles for 'TotalWorkingYears'
quartiles = data['TotalWorkingYears'].quantile([0.25, 0.5, 0.75])

# Print the quartiles
print("Quartiles for TotalWorkingYears:")
print(quartiles)

# Visualize the distribution with quartiles highlighted
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalWorkingYears'], kde=True)
plt.axvline(quartiles[0.25], color='r', linestyle='--', label='Q1')
plt.axvline(quartiles[0.50], color='g', linestyle='--', label='Q2 (Median)')
plt.axvline(quartiles[0.75], color='b', linestyle='--', label='Q3')
plt.title('Distribution of Total Working Years with Quartiles')
plt.xlabel('Total Working Years')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
